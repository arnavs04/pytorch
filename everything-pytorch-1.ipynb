One of the most common erros is shape errors
2 Main Rules for matrix multiplcation
*   Inner dimensions in shape must match
    *   (3) @ (3) will work
    *   (3,2) @ (3,2) won't work
    *   (3,2) @ (2,3) will work
    *   (3,2) @ (2,4) will work

*   Outer dimensions in shape form the resultant matrix
    *   (3) @ (3) gives 1,1
    *   (3,2) @ (2,3) gives 3,3
    *   (3,2) @ (2,4) gives 3,4

```python
# Linear Regression Example using Labelled Data
weight = 0.7
bias = 0.3

# Creating the Data
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim = 1)
y = weight * X + bias
```

Splitting Data!!!


1.   Training Set (model learns patterns from here) (around 60-70%) **(mandatory)**
2.   Validation Set (tune model patterns- hyperparameters) (10-20%) (often but not always)
3.   Test Set (to check if the model is ready for the real world usage) (10-20%) **(always)**






```python
# So let's split

train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
```

```python
# Visualisation

def plot_predictions(train_data = X_train,
                     train_labels = y_train,
                     test_data = X_test,
                     test_labels = y_test,
                     predictions = None):

    """Ploting all the data"""

    plt.figure(figsize = (10,7))

    # Plot training data in blue

    plt.scatter(train_data, train_labels, c = "b", s = 4, label = "Training data")
    plt.scatter(test_data, test_labels, c = "g", s = 4, label = "Testing data")

    if predictions is not None:
        plt.scatter(test_data, predictions, c = "r", s = 4, label = "Predictions")

    plt.legend(prop = {"size" : 14})

    plt.show()
```

```python
plot_predictions()
```

### Creating PyTorch Model

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1, # start with random weight at first
                                                requires_grad = True, # needs to be updated with gradient descent
                                                dtype = torch.float)) # pytorch loves data type float32
        self.bias = nn.Parameter(torch.randn(1,
                                             requires_grad = True,
                                             dtype = torch.float))
        # Basically
        # self.linear_layer = torch.nn.Linear(in_features = 1, out_features = 1)
    def forward(self, x: torch.Tensor) -> torch.Tensor: # forward propagation
            return (self.weights * x + self.bias)
```

**torch.nn** - contains all the building blocks for computational graphs (neural network)

**torch.nn.Parameter** - what parameters should our model try and learn, often a PyTorch layer from torch.nn will set these for us

**torch.nn.Module** - base class from all neural network modules if you subclass it, you should overwrite forward()

**torch.optim** - this where the optimizers in PyTorch live, they will help with gradient descent

**def forward** - all nn.Module subclasses require you to overwrite forward(), this method defines what happens in the forward computation

**torch.utils.data.Dataset** - represents a map between key (label) and sample (features) pairs of your data

**torch.utils.data.DataLoader** - creates a python iterable over a torch dataset (allows you to iterate over your data)

### Training a Model

```python
# Loss function help represent how far off the predictions are from the labelled output
# Minimizing them is the whole goal

# Optimizer looks at the loss and adjusts the model's parameters (e.g. weights & biases)

list(model_0.parameters())
```

```python
model_0.state_dict()
```

So we need a
1.   Training Loop
2.   Testing Loop

Which loss function and which optimizer we use is only problem specific. With experience, you'll get an idea of what works ad what doesn't.

```python
# Now we need a loss function
loss_fn = nn.L1Loss()

# Setting up an optimizer
optimizer = torch.optim.SGD(params = model_0.parameters(), # parameters
                            lr = 0.01) # learning rate
```

Training Loop!!!


1.   Loop through the data
2.   Forward Propagation (making predictions on data)
3.   Calculating the loss (compare forward pass predictions to labels)
4.   Optimizer Zero Grad (Set the gradients of all parameters to zero initially)
5.   Backward Propagation (move backwards throught the network to calculate the gradients of each of the parameters of our model with respect to the loss)
6.   Gradient Descent (use optimizer to adjust our model's parameters to try and improve the loss)





```python
list(model_0.parameters())
```

```python
torch.manual_seed(42)

# An epoch is one loop through the data (a hyperparameter)
epochs = 200

# Keep track of useful values
epoch_count = []
loss_values = []
test_loss_values = []

# Putting data on target device (device agnostic code for data)
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)


# 1. Loop through the data
for epoch in range(epochs):

    # Set the model to training mode
    model_0.train()
    # train mode in PyTorch sets all parameters that require gradients to require gradients to

    # 2. Forward Propagation
    y_pred = model_0(X_train)

    # 3. Loss Calculation
    loss = loss_fn(y_pred, y_train)

    # 4. Optimizer Zero Grad
    optimizer.zero_grad()

    # 5. Back propagation
    loss.backward() # with requires_grad = True

    # 6. Gradient Descent
    optimizer.step()

    # Testing
    model_0.eval()
    # Turns off diff settings not needed for evaluation/testing like dropout/batchnorm
    with torch.inference_mode(): # turns off gradient tracking & a couple more things behind the scenes
    # with torch.no_grad(): # You may also see torch.no_grad in older pytorch code
        # 1. Forward pass
        test_pred = model_0(X_test)

        # 2. Calculate loss
        test_loss = loss_fn(test_pred, y_test)
    if epoch % 10 == 0:
        epoch_count.append(epoch)
        loss_values.append(loss)
        test_loss_values.append(test_loss)

        # Printing losses with epoch & their own parameters
        print(f"Epoch: {epoch} | Loss: {loss} | Test loss: {test_loss}")

        print(f"Parameters: {model_0.state_dict()}")
```

```python
# Constructing the class with init & forward pass

class CircleModelV0(nn.Module):
    def __init__(self):
        super().__init__()
        # Creating the layers
        self.layer_1 = nn.Linear(in_features = 2, out_features = 5)
        self.layer_2 = nn.Linear(in_features = 5, out_features = 1)

        # You can also do
        # self.two_linear_layers =  = nn.Sequential(
        #                               nn.Linear(in_features = 2, out_features = 5),
        #                               nn.Linear(in_features = 5, out_features = 1))
        # and below then return self.two_linear_layers(x)

    def forward(self, x):
        return (self.layer_2(self.layer_1(x))) # x -> layer 1 -> layer 2

        # instead of just return ..
        # we do x = self.layer_1(x)
        # and x = self.layer_2(x)
        # return x
```

### Training & Testing Loop for Model 1

```python
# Function for Training - train_step()

def train_step(model: nn.Module,
               data_loader: torch.utils.data.DataLoader,
               loss_fn: nn.Module,
               optimizer: torch.optim.Optimizer,
               accuracy_fn,
               device: torch.device = device):

    train_loss, train_acc = 0, 0

    # Training Mode
    model.train()

    for batch, (X, y) in enumerate(data_loader):

        # Putting data on target device
        X, y = X.to(device), y.to(device)

        y_pred = model(X) # Forward Pass

        loss = loss_fn(y_pred, y) # Calculating Loss
        train_loss += loss # Accumulating Training Loss
        train_acc += accuracy_fn(y_true = y,
                                 y_pred = y_pred.argmax(dim = 1)) # from logits -> prediction label

        optimizer.zero_grad() # Optimizer Zero Grad

        loss.backward() # Backward Propagation

        optimizer.step() # update the model's parameters once *per batch*

    train_loss /= len(data_loader)
    train_acc /= len(data_loader)

    print(f"\nTrain Loss: {train_loss:.2f} | Train Acc: {train_acc:.2f}")
```

```python
# Function for Testing - test_step()

def test_step(model: nn.Module,
               data_loader: torch.utils.data.DataLoader,
               loss_fn: nn.Module,
               accuracy_fn,
               device: torch.device = device):

    test_loss, test_acc = 0, 0

    model.eval() # Evaluation Mode

    with torch.inference_mode():

        # Looping through Data
        for X_test, y_test in data_loader:
            X_test, y_test = X_test.to(device), y_test.to(device)

            test_pred = model(X_test)

            test_loss += loss_fn(test_pred, y_test)
            test_acc += accuracy_fn(y_true = y_test,
                                    y_pred = test_pred.argmax(dim = 1))

        test_loss /= len(data_loader)
        test_acc /= len(data_loader)

    print(f"\nTest Loss: {test_loss:.2f} | Test Acc: {test_acc:.2f}")
```

```python
# Training & Testing Loop

torch.manual_seed(42)

# Measure Time
train_time_start_on_gpu = timer()

# Set Epochs
epochs = 3

# Create an optimization and evaluation loop using train_step() and test_step()
for epoch in tqdm(range(epochs)):
    print(f"Epoch: {epoch}\n------")

    train_step(model = model_1,
               data_loader = train_dataloader,
               loss_fn = loss_fn,
               optimizer = optimizer,
               accuracy_fn = accuracy_fn,
               device = device)

    test_step(model = model_1,
              data_loader = test_dataloader,
              loss_fn = loss_fn,
              accuracy_fn = accuracy_fn,
              device = device)

train_time_end_on_gpu = timer()
total_train_time_model_1 = print_train_time(start = train_time_start_on_gpu,
                                            end = train_time_end_on_gpu,
                                            device = device)
```

## Convolutional Neural Network Model

### Building a CNN

```python
# This architecture replicates TinyVGG

class FashioMNISTModelV2(nn.Module):
    def __init__(self,
                 input_shape: int,
                 hidden_units: int,
                 output_shape: int):
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels = input_shape,
                      out_channels = hidden_units,
                      kernel_size = 3,
                      stride = 1,
                      padding = 1), # "Hyperparameters"
            nn.ReLU(),
            nn.Conv2d(in_channels = hidden_units,
                      out_channels = hidden_units,
                      kernel_size = 3,
                      stride = 1,
                      padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2)

        )

        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(in_channels = hidden_units,
                      out_channels = hidden_units,
                      kernel_size = 3,
                      stride = 1,
                      padding = 1),
            nn.ReLU(),
            nn.Conv2d(in_channels = hidden_units,
                      out_channels = hidden_units,
                      kernel_size = 3,
                      stride = 1,
                      padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2),

        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            # Where did this in_features shape come from?
            # It's because each layer of our network compresses and
            # changes the shape of our inputs data.
            nn.Linear(in_features = hidden_units*7*7,
                      # Calculated by putting x.shape after every block in forward func
                      out_features = output_shape)
        )

    def forward(self, x: torch.Tensor):
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.classifier(x)
        return x

```

## Making Predictions with Best Model

```python
def make_predictions(model: nn.Module,
                     data: list,
                     device: torch.device = device):
    pred_probs = []
    model.to(device)
    model.eval()

    with torch.inference_mode():
        for sample in data:
            # Prepare the sample (add a batch dimension and pass to target device)
            sample = torch.unsqueeze(sample, dim = 0).to(device)

            # Forward pass (model outputs raw logits)
            pred_logit = model(sample)

            # Get prediction probability (logit -> prediction probability)
            pred_prob = torch.softmax(pred_logit.squeeze(), dim = 0)

            # Get pred_prob off the GPU for further calculations
            pred_probs.append(pred_prob.cpu())

    # Stack the pred_probs to turn list into a tensor
    return torch.stack(pred_probs)
```
